```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt

data_noiseLevel = loadmat("ResultsLog/results_noiseLevel.mat")
arr_sig_NR = data_noiseLevel["arr_sig_NR"]
error_SINDy_manual = data_noiseLevel["error_SINDy_manual"]
TPR_SINDy_manual = data_noiseLevel["TPR_SINDy_manual"]
error_bumpWSINDy_manual = data_noiseLevel["error_bumpWSINDy_manual"]
TPR_bumpWSINDy_manual = data_noiseLevel["TPR_bumpWSINDy_manual"]
error_FourierWSINDy_manual = data_noiseLevel["error_FourierWSINDy_manual"]
TPR_FourierWSINDy_manual = data_noiseLevel["TPR_FourierWSINDy_manual"]
```

### Prescribed $\lambda$

Low noise

```python
inds = range(0,5)
# error norm
plt.figure()
plt.loglog(arr_sig_NR[0,inds].flatten(),error_SINDy_manual[0,inds].flatten())
plt.loglog(arr_sig_NR[0,inds].flatten(),error_bumpWSINDy_manual[0,inds].flatten())
plt.loglog(arr_sig_NR[0,inds].flatten(),error_FourierWSINDy_manual[0,inds].flatten())
plt.xlabel("Noise Ratio")
plt.ylabel("Relative Error Norm")
plt.legend(["SINDy","Weak SINDy (Bump)","Weak SINDy (Fourier)"])

# true positive ratio
plt.figure()
plt.semilogx(arr_sig_NR[0,inds].flatten(),TPR_SINDy_manual[0,inds].flatten())
plt.semilogx(arr_sig_NR[0,inds].flatten(),TPR_bumpWSINDy_manual[0,inds].flatten())
plt.semilogx(arr_sig_NR[0,inds].flatten(),TPR_FourierWSINDy_manual[0,inds].flatten())
plt.xlabel("Noise Ratio")
plt.ylabel("TPR")
plt.legend(["SINDy","Weak SINDy (Bump)","Weak SINDy (Fourier)"])
```

High noise

```python
inds = range(5,13)
# error norm
plt.figure()
plt.loglog(arr_sig_NR[0,inds].flatten(),error_SINDy_manual[0,inds].flatten())
plt.loglog(arr_sig_NR[0,inds].flatten(),error_bumpWSINDy_manual[0,inds].flatten())
plt.loglog(arr_sig_NR[0,inds].flatten(),error_FourierWSINDy_manual[0,inds].flatten())
plt.xlabel("Noise Ratio")
plt.ylabel("Relative Error Norm")
plt.legend(["SINDy","Weak SINDy (Bump)","Weak SINDy (Fourier)"])

# true positive ratio
plt.figure()
plt.loglog(arr_sig_NR[0,inds].flatten(),TPR_SINDy_manual[0,inds].flatten())
plt.loglog(arr_sig_NR[0,inds].flatten(),TPR_bumpWSINDy_manual[0,inds].flatten())
plt.loglog(arr_sig_NR[0,inds].flatten(),TPR_FourierWSINDy_manual[0,inds].flatten())
plt.xlabel("Noise Ratio")
plt.ylabel("TPR")
plt.legend(["SINDy","Weak SINDy (Bump)","Weak SINDy (Fourier)"])
```
