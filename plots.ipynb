```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
```

### Lorenz - Plots

```python
# load data
data = loadmat("ResultsLog/results_noiseLevel_Lorenz.mat")
arr_sig_NR = data["arr_sig_NR"].flatten()

# extract methods and colors
methods = [
    ("SINDy", "errorMean_SINDy_Lorenz", "errorStD_SINDy_Lorenz", "TPRMean_SINDy_Lorenz", "TPRStD_SINDy_Lorenz", "tab:red"),
    ("Bump WSINDy", "errorMean_bumpWSINDy_Lorenz", "errorStD_bumpWSINDy_Lorenz", "TPRMean_bumpWSINDy_Lorenz", "TPRStD_bumpWSINDy_Lorenz", "tab:blue"),
    ("Fourier WSINDy (Trapezoid)", "errorMean_FourierWSINDy_Lorenz", "errorStD_FourierWSINDy_Lorenz", "TPRMean_FourierWSINDy_Lorenz", "TPRStD_FourierWSINDy_Lorenz", "tab:green"),
    ("Fourier WSINDy (FFT)", "errorMean_FourierWSINDyFFT_Lorenz", "errorStD_FourierWSINDyFFT_Lorenz", "TPRMean_FourierWSINDyFFT_Lorenz", "TPRStD_FourierWSINDyFFT_Lorenz", "tab:purple"),
]

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# error plot
ax1 = axs[0]
for name, err_mean_key, err_std_key, _, _, color in methods:
    mean = data[err_mean_key].flatten()
    std = data[err_std_key].flatten()
    ax1.plot(arr_sig_NR, mean, label=name, color=color)
    ax1.fill_between(arr_sig_NR, mean - std, mean + std, color=color, alpha=0.2)

ax1.set_title("Coefficient Error vs. Noise Ratio", fontsize=14)
ax1.set_xlabel("Noise Ratio", fontsize=12)
ax1.set_ylabel("Coefficient Error", fontsize=12)
ax1.grid(True, linestyle="--", alpha=0.5)
ax1.legend()

# TPR plot
ax2 = axs[1]
for name, _, _, tpr_mean_key, tpr_std_key, color in methods:
    mean = data[tpr_mean_key].flatten()
    std = data[tpr_std_key].flatten()
    ax2.plot(arr_sig_NR, mean, label=name, color=color)
    ax2.fill_between(arr_sig_NR, mean - std, mean + std, color=color, alpha=0.2)

ax2.set_title("True Positive Ratio vs. Noise Ratio", fontsize=14)
ax2.set_xlabel("Noise Ratio", fontsize=12)
ax2.set_ylabel("True Positive Ratio", fontsize=12)
ax2.grid(True, linestyle="--", alpha=0.5)
ax2.legend()

plt.tight_layout()
plt.show()
```

### Lorenz - Speed Comparisons

```python
time_SINDy = data["time_SINDy"]
time_bumpWSINDy = data["time_bumpWSINDy"]
time_FourierWSINDy = data["time_FourierWSINDy"]
time_FourierWSINDyFFT = data["time_FourierWSINDyFFT"]

print(f"SINDy: {time_SINDy[0,0]}s")
print(f"Bump Weak SINDy: {time_bumpWSINDy[0,0]}s")
print(f"Fourier Weak SINDy: {time_FourierWSINDy[0,0]}s")
print(f"Fourier Weak SINDy with FFT: {time_FourierWSINDyFFT[0,0]}s")
```
